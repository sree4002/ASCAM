```python
from ultralytics import YOLO
import os

# Load a YOLOv8 model; we use a pretrained model as a base, such as YOLOv8n or YOLOv8s.
model = YOLO("yolov8n.pt")  # Choose a YOLOv8 model variant (e.g., yolov8n.pt for Nano, yolov8s.pt for Small, etc.)

# Train the model
results = model.train(
    data=os.path.join("/content/gdrive/MyDrive/ASCAM/data.yaml"),  # Path to the data.yaml file
    epochs=500,                         # Number of epochs to train
    imgsz=640,                         # Image size for training
    batch=16,                          # Batch size; adjust based on your hardware
    name="axonal_swelling_detector",   # Name for the training run
    project="axonal_detection_project" # Directory to save training results
)
```

```python
# Save the trained model
model_path = os.path.join("/content/gdrive/MyDrive/ASCAM", "weights.pt")
model.save(model_path)
print(f"Model saved to {model_path}")
```

```python
from google.colab import drive
drive.mount('/content/drive', force_remount=True)   # FORCE the mount

import os
from ultralytics import YOLO
from IPython.display import Image, display
import glob
import sys
from pathlib import Path


# â”€â”€â”€ 1) Verify the model folder and list its contents â”€â”€â”€
BASE = Path("/content/drive/MyDrive/ASCAM/Object_Detection_Model")
if not BASE.exists():
    print(f"ERROR: folder not found: {BASE}", file=sys.stderr)
    raise SystemExit

print("Contents of your model folder:")
for f in sorted(os.listdir(BASE)):
    print(" ", f)

# â”€â”€â”€ 2) Pick the correct weights filename â”€â”€â”€
#    Adjust this if your checkpoint is named something else!
weights_file = BASE / "weights.pt"
if not weights_file.exists():
    print(f"ERROR: weights.pt not found at {weights_file}", file=sys.stderr)
    raise SystemExit

# â”€â”€â”€ 3) Load the model and run val with plots â”€â”€â”€
model = YOLO(str(weights_file))

metrics = model.val(
    plots=True,
    conf=0.001,
    iou=0.50,
    max_det=1000,
    verbose=False
)

print("\nValidation Metrics:")
print(f"  Precision:     {metrics.box.p[0]:.4f}")
print(f"  Recall:        {metrics.box.r[0]:.4f}")
print(f"  mAP@0.5:       {metrics.box.map50:.4f}")
print(f"  mAP@0.5â€“0.95:  {metrics.box.map:.4f}")

# â”€â”€â”€ 4) Display the saved plots â”€â”€â”€
run_dirs = sorted(glob.glob("runs/val/*"))
if not run_dirs:
    print("No runs/val folder found!", file=sys.stderr)
    raise SystemExit

latest = run_dirs[-1]
print("\nShowing plots from:", latest)

for name in ("confusion_matrix.png", "pr_curve.png"):
    path = Path(latest) / name
    if path.exists():
        display(Image(str(path)))
    else:
        print(f"Plot not found: {path}", file=sys.stderr)
```
